```python
# Web3.eth API Setup

from web3 import Web3
from dotenv import load_dotenv, set_key
import os
import time
import json

# Web3 API Setup
load_dotenv('../.env')
w3 = Web3(Web3.HTTPProvider(os.getenv('HTTPProvider')))
```

```python
# Get block data from Web3
block_hash_w3 = w3.eth.get_block('latest')['hash']
block_num_w3 = w3.eth.get_block('latest')['number']

print(block_hash_w3.hex())
print(block_num_w3)
```

```python
# Block Info
block_info_w3 = w3.eth.get_block('latest')

for key, value in block_info_w3.items():
    if isinstance(value, bytes):
        value = value.hex()
    if type(block_info_w3[key]) is list:
        for i in block_info_w3[key]:
            if isinstance(i, bytes):
                i = i.hex()
                print(f"    {i}")
            else:
                print(f"    {i}")
    else:
        print(f'{key}: {value}')
```

```python
# Block Info with Full Transactions Data
block_info_transactions_w3 = w3.eth.get_block('latest', full_transactions=True)

for key, value in block_info_transactions_w3.items():
    if isinstance(value, bytes):
        value = value.hex()
    if type(block_info_transactions_w3[key]) is list:
        for i in block_info_transactions_w3[key]:
            if isinstance(i, bytes):
                i = i.hex()
                print(f"    {i}")
            else:
                print(f"    {i}")
    else:
        print(f'{key}: {value}')
```

```python
block_data = {}
for key, value in block_info_transactions_w3.items():
    if isinstance(value, bytes):
        value = value.hex()
    if isinstance(block_info_transactions_w3[key], list):
        procesed_value = []
        for i in block_info_transactions_w3[key]:
            if isinstance(i, bytes):
                i = i.hex()
                procesed_value.append(i)
            else:
                procesed_value.append(i)
        block_data[key] = value
    else:
        block_data[key] = value

print(block_data)
```

```python
# Block Transactions Data Only
block_transactions_w3 = w3.eth.get_block('latest', full_transactions=True)['transactions']

for transaction in block_transactions_w3:
    print(transaction)
```

```python
# Block Single Transactions Data Only
block_transactions_w3 = w3.eth.get_block('latest', full_transactions=True)['transactions']

for key, value in block_transactions_w3[0].items():
    if isinstance(value, bytes):
        value = value.hex()
        print(f'{key}: {value}')
    else:
        print(f'{key}: {value}')
```
